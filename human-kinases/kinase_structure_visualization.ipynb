# Notebook to create a 3D figure exemplifying the typical kinase fold

Use opencadd KLIFS module

```python
from opencadd.structure.pocket import PocketKlifs, PocketViewer
```

## Load pocket
As an example EGFR structure [2ito](https://klifs.net/details.php?structure_id=783) is taken.

```python
klifs_id = 783
```

```python
pocket = PocketKlifs.from_structure_klifs_id(klifs_id)
```

## Visualization

This is using nglview under the hood

* add pocket
* add ligand
* draw
* export

```python
v = PocketViewer()
v.add_pocket(pocket, ligand_expo_id="IRE", show_pocket_center=False)
v.viewer.add_representation("licorice", selection="IRE")
v.viewer.center(selection="IRE")
v.viewer
```

```python
v.viewer.render_image();
```

```python
v.viewer._display_image()
```

```python
v.viewer.download_image(filename='kinaseStructure.png', transparent=True)
```
